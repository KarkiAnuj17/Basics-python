```python
class Employee:
    count = 0 
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.count += 1

    #instances method    
    def employee_detail(self):
        print(self.name, self.salary)
        
    @classmethod
    def total_no_employee(cls):
        return cls.count

    @staticmethod
    def valid_salary(salary):
        if salary > 0:
            return True
        else:
            return False


s1 = Employee("Ram", 20000)
s2 = Employee("hari", 30000)

s1.employee_detail()
s2.employee_detail()

print(Employee.valid_salary(10000)) 
print(Employee.valid_salary(-10000)) 

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)   #super used to access from the parent class
        self.department = department

    def display_info(self):
        print(self.name, self.salary, self.department)

m1 = Manager("Sita", 5000, "HR")
m1.display_info()
print(Employee.total_no_employee())
```

```python
class Message:
    def __init__(self, from_number, time_arrived, text_of_SMS):
        self.has_been_viewed = False
        self.from_number = from_number
        self.time_arrived = time_arrived
        self.text_of_SMS = text_of_SMS

    def get_details(self):
        return self.has_been_viewed, self.from_number, self.time_arrived, self.text_of_SMS

    def __repr__(self):
        return f"({self.has_been_viewed}, {self.from_number}, {self.time_arrived}, {self.text_of_SMS})"


class SMS_store:
    def __init__(self):
        self.messages = []

    def add_new_arrival(self, from_number, time_arrived, text_of_SMS):
        new_message = Message(from_number, time_arrived, text_of_SMS)
        self.messages.append(new_message)
        return self.messages

    def message_count(self):
        return len(self.messages)

    def get_unread_indexes(self):
        unread_indexes = []
        for i, message in enumerate(self.messages):  
            if message.has_been_viewed == False:
                unread_indexes.append(i)
        return unread_indexes

    def get_message(self, i):
        if 0 <= i < len(self.messages):
            message = self.messages[i]
            message.has_been_viewed = True
            return message.get_details()
        return None

    def delete(self, i):         
        del self.messages[i]

    def clear(self):
        self.messages.clear()
        
my_inbox = SMS_store()
print(my_inbox.add_new_arrival(9999999, 2, "wewewe"))
print(my_inbox.add_new_arrival(4499944, 4, "wewfghjjfewe"))
print(my_inbox.add_new_arrival(9999999, 5, "wewdfddfgfewesfdfs"))
print(my_inbox.message_count())
print(my_inbox.get_unread_indexes()) 
print(my_inbox.get_message(0))         
my_inbox.delete(0)
print(my_inbox.message_count())
my_inbox.clear()
print(my_inbox.message_count())
```

```python
class Patient:
    count = 0
    auto_id = 0  

    def __init__(self, name, age, gender, medical_history):
        if age <= 0:
            raise ValueError("Age must be greater than 0.")
        self.name = name
        self.age = age
        self.gender = gender
        self.patient_id = Patient.auto_id
        Patient.auto_id += 1
        self.medical_history = medical_history
        Patient.count += 1

    def get_summary(self):
        return self.name, self.age, self.gender, self.patient_id, self.medical_history

    def add_medical_record(self, condition):
        self.medical_history.append(condition)


p1 = Patient("Ram", 22, "Male", ["Diabetes", "Blood Pressure"])
print(p1.get_summary())
p1.add_medical_record("Asthma")
print(p1.get_summary())


class Inpatient(Patient):
    def __init__(self, name, age, gender, medical_history, admission_date, room_number):
        super().__init__(name, age, gender, medical_history)
        self.admission_date = admission_date
        self.room_number = room_number

    def discharge(self):
        self.admission_date = None
        self.room_number = None

    def get_summary_info(self):
        return (self.name, self.age, self.gender, self.patient_id,
                self.medical_history, self.admission_date, self.room_number)


p2 = Inpatient("Sita", 35, "Female", ["Asthma", "Thyroid"], "2024-06-10", 202)
print(p2.get_summary_info())
p2.discharge()
print(p2.get_summary_info())


class Outpatient(Patient):
    def __init__(self, name, age, gender, medical_history, last_visit_date, doctor_consulted):
        super().__init__(name, age, gender, medical_history)
        self.last_visit_date = last_visit_date
        self.doctor_consulted = doctor_consulted

    def get_summary_info(self):
        return (self.name, self.age, self.gender, self.patient_id,
                self.medical_history, self.last_visit_date, self.doctor_consulted)

    def schedule_follow_up(self, new_date):
        self.last_visit_date = new_date


p3 = Outpatient("Hari", 30, "Male", ["Blood Sugar", "Thyroid"], "2024-05-10", "Dr.Joe")
print(p3.get_summary_info())
p3.schedule_follow_up("2024-05-30")
print(p3.get_summary_info())

print(Patient.count)
```
